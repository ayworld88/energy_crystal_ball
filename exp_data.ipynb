```python
# load dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# read file
state_data = pd.read_csv('state_prod.csv')
state_data.head()
```

```python
# view columns/headers
state_data.columns
```

```python
# drop column
state_data = state_data.drop(columns =["Data_Status"])
# rename column
state_data = state_data.rename(columns={'StateCode': 'State'})
state_data
```

```python
# restructure dataframe
melt_data = pd.melt(state_data, id_vars=['State', 'MSN'])
melt_data
```

```python
melt_data = melt_data.rename(columns={'variable': 'Year'})
melt_data
```

```python
# Build a scatter plot for each data type
plt.scatter(melt_data["Year"], melt_data["value"], marker="o", facecolors="blue", edgecolors="black")

# Incorporate the other graph properties
plt.title("Prod vs. Year")
plt.ylabel("Production (Billion Btu)")
plt.xlabel("Year")

plt.figure(figsize=(40,20))


# Show plot
plt.show()

```

```python
df_with_dummies = pd.get_dummies(melt_data, columns = ["MSN", "State"])
df_with_dummies
```

```python
df_with_dummies.columns
```

```python
# Import dependencies
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
# Assign X (data) and y (target)
X = df_with_dummies[['Year', 'MSN_CLPRB', 'MSN_CLPRK', 'MSN_CLPRP', 'MSN_COPRK',
       'MSN_EMFDB', 'MSN_ENPRP', 'MSN_GETCB', 'MSN_HYTCB', 'MSN_NGMPB',
       'MSN_NGMPK', 'MSN_NGMPP', 'MSN_NUETB', 'MSN_PAPRB', 'MSN_PAPRP',
       'MSN_REPRB', 'MSN_ROPRB', 'MSN_SOTCB', 'MSN_TEPRB', 'MSN_TETCB',
       'MSN_WDEXB', 'MSN_WDPRB', 'MSN_WDTCB', 'MSN_WSTCB', 'MSN_WYTCB',
       'State_AK', 'State_AL', 'State_AR', 'State_AZ', 'State_CA', 'State_CO',
       'State_CT', 'State_DC', 'State_DE', 'State_FL', 'State_GA', 'State_HI',
       'State_IA', 'State_ID', 'State_IL', 'State_IN', 'State_KS', 'State_KY',
       'State_LA', 'State_MA', 'State_MD', 'State_ME', 'State_MI', 'State_MN',
       'State_MO', 'State_MS', 'State_MT', 'State_NC', 'State_ND', 'State_NE',
       'State_NH', 'State_NJ', 'State_NM', 'State_NV', 'State_NY', 'State_OH',
       'State_OK', 'State_OR', 'State_PA', 'State_RI', 'State_SC', 'State_SD',
       'State_TN', 'State_TX', 'State_US', 'State_UT', 'State_VA', 'State_VT',
       'State_WA', 'State_WI', 'State_WV', 'State_WY']]
y = df_with_dummies["value"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Split the data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScater model and fit it to the training data
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Create a LinearRegression model and fit it to the scaled training data

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)
```

```python
# Make predictions using the X_test_scaled data
# Plot y_test_scaled vs y_test_scaled
# Scatter plot y_test_scaled vs predictions

predictions = model.predict(X_test_scaled)
model.fit(X_train_scaled, y_train_scaled)
plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train_scaled, c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), model.predict(X_test_scaled) - y_test_scaled, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```

```python
# Used X_test_scaled, y_test_scaled, and model.predict(X_test_scaled) to calculate MSE and R2

from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = model.score(X_test_scaled, y_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")
```
