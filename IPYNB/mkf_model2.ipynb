```python

# load dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
state_energy = pd.read_csv("state_prod.csv")
state_energy.head()
```

```python
# drop column
state_energy = state_energy.drop(columns =["Data_Status"])
# rename column
state_energy = state_energy.rename(columns={'StateCode': 'State'})
state_energy.head()
```

```python
#reset datafram index and remove uneccessary columns 
state_energy.set_index('MSN', inplace=True)
state = state_energy.drop(["CLPRK","CLPRP","COPRK","EMFDB", "ENPRP","GETCB","HYTCB",
                  "NGMPK","NGMPP","NUETB","PAPRP","SOTCB", "TEPRB","TETCB","WDEXB", "WDTCB", "WSTCB", "WYTCB"])
state.head()
```
